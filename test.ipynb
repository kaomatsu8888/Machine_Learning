```python
from sklearn import datasets
iris = datasets.load_iris()
print(iris)
```

```python
print("特徴量の名前=", iris.feature_names)
print("分類の名前=", iris.target_names)
print("分類の値=", iris.target)
```

```python
# データをデータフレームにいれる
import pandas as pd
df = pd.DataFrame(iris.data)
df.head()
```

```python
# 特徴量の名前を設定
df.columns = iris.feature_names # 特徴量の名前を設定
df["target"] = iris.target # 分類の値を設定
df.head() # 先頭を表示
```

```python
# ヒストグラムで描写(3種類の品種を色分け)
%matplotlib inline
import matplotlib.pyplot as plt # for plotting

# 3種類の品種を、別々のデータフレームに分ける
# targetの値が0ならdf0、1ならdf1、2ならdf2に入れる
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]

# 3種類の品種を、別々のヒストグラムとして描写
# 「がくの幅」で、ヒストグラムを描写
plt.figure(figsize=(5,5)) # 図のサイズを指定
xx = "sepal width (cm)" # xxに「がくの幅」を設定
df0[xx].plot.hist(color="b", alpha=0.5) # df0のヒストグラムを描写
df1[xx].plot.hist(color="r", alpha=0.5) # df1のヒストグラムを描写
df2[xx].plot.hist(color="g", alpha=0.5) # df2のヒストグラムを描写
plt.xlabel(xx) # x軸のラベルを表示
plt.show() # 図を表示
```

```python
# 「がくの幅」と「がくの長さ」の関係を散布図で描写
xx = "sepal width (cm)" # xxに「がくの幅」を設定
yy = "sepal length (cm)" # yyに「がくの長さ」を設定
plt.figure(figsize=(5,5)) # 図のサイズを指定
plt.scatter(df0[xx], df0[yy], color="b", alpha=0.5) # df0の散布図を描写
plt.scatter(df1[xx], df1[yy], color="r", alpha=0.5) # df1の散布図を描写
plt.scatter(df2[xx], df2[yy], color="g", alpha=0.5) # df2の散布図を描写
plt.xlabel(xx) # x軸のラベルを表示
plt.ylabel(yy) # y軸のラベルを表示
plt.grid() # グリッドを表示
plt.show() # 図を表示


```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

xx = "sepal width (cm)" # xxに「がくの幅」を設定
yy = "sepal length (cm)" # yyに「がくの長さ」を設定
zz = "petal length (cm)" # zzに「花弁の長さ」を設定

fig = plt.figure(figsize=(5,5)) # 図のサイズを指定
ax = fig.add_subplot(111, projection='3d') # 3D軸を作成

# df0, df1, df2 のデータを用いて散布図を描画
# 注: ここでは df0, df1, df2 が定義されていると仮定しています
ax.scatter(df0[xx], df0[yy], df0[zz], color="b") # df0の散布図を描写
ax.scatter(df1[xx], df1[yy], df1[zz], color="r") # df1の散布図を描写
ax.scatter(df2[xx], df2[yy], df2[zz], color="g") # df2の散布図を描写

ax.set_xlabel(xx) # x軸のラベルを設定
ax.set_ylabel(yy) # y軸のラベルを設定
ax.set_zlabel(zz) # z軸のラベルを設定

ax.view_init(0,240) # 図の角度を設定(0,240

plt.show() # 図を表示
```

```python
# 塊数が2つのデータを作成
from sklearn.datasets import make_blobs
import pandas as pd

# データを作成
X, y = make_blobs(random_state=3, # 乱数を固定
                n_features=2, # 特徴量を2つ
                centers=2, # クラスターの数を2つ
                cluster_std=1, # クラスターのばらつき度合い
                n_samples=300)# データの個数を300個に設定
df = pd.DataFrame(X) # データフレームに変換
df["target"] = y # 分類の値を設定
df.head() # 先頭を表示
```

```python
%matplotlib inline
import matplotlib.pyplot as plt # for plotting

# 分類によって、別々のデータフレームに分ける
df0 = df[df.target==0]
df1 = df[df.target==1]
# 分類0は青、分類1は赤で、散布図を描写
plt.figure(figsize=(5,5)) # 図のサイズを指定
plt.scatter(df0[0], df0[1], color="b", alpha=0.5) # 分類0の散布図を描写
plt.scatter(df1[0], df1[1], color="r", alpha=0.5) # 分類1の散布図を描写
plt.grid() # グリッドを表示
plt.show() # 図を表示
```

```python
# make_regressionを使って,回帰用のデータセットを作成.ノイズ10でxが0のときyが100になるように設定
from sklearn.datasets import make_regression
X, y = make_regression(
                n_features=1, # 特徴量を1つに設定
                n_samples=300, # データの個数を300個に設定
                noise=10, # ノイズの大きさを10に設定
                bias =100, # xが0のときyが100になるように設定
                random_state=3) # 乱数を固定

# データフレームを作り「特徴量0」と「y」で、散布図を描写
df = pd.DataFrame(X) # データフレームに変換
plt.figure(figsize=(5,5)) # 図のサイズを指定
plt.scatter(df[0], y, color="b", alpha=0.5) # 散布図を描写
plt.grid() # グリッドを表示
plt.show() # 図を表示
```
